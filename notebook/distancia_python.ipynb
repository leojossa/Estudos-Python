```python
import  locale
import warnings
import pandas as pd

warnings.filterwarnings('ignore')
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
```

```python
df = pd.read_csv(r'C:\Users\kaleo\Documents\dados_exemplos_v1.csv', dtype = {'CEP_CLIENTE':'str','CEP_VENDEDOR':'str'},sep=';')
```

```python
df_cep_cli = df['CEP_CLIENTE']
df_cep_cli = pd.DataFrame(df_cep_cli)
df_cep_cli.drop_duplicates(inplace = True)
df_cep_cli.head()
```

```python
df_cep_ven = df['CEP_VENDEDOR']
df_cep_ven = pd.DataFrame(df_cep_ven)
df_cep_ven.drop_duplicates(inplace = True)
df_cep_ven.head()
```

```python
print('CEPS unicos vendedores',df_cep_ven.count())
print('CEPS unicos clientes',df_cep_cli.count())
```

```python
import urllib.request
import json

for idx, row in df_cep_ven.iterrows():
    cep = row.CEP_VENDEDOR
    url = 'https://viacep.com.br/ws/%s/json/' % cep
    headers = {'User-Agent':'Teste'}
    requisicao = urllib.request.Request(url, headers=headers, method='GET')
    cliente = urllib.request.urlopen(requisicao)
    conteudo = cliente.read().decode('utf-8')
    endereco = json.loads(conteudo)
    cliente.close()
    print(cep)

    if 'cep' in endereco:
        df_cep_ven.at[idx, 'LOGRADOURO_VENDEDOR'] = endereco['logradouro']
        df_cep_ven.at[idx, 'BAIRRO_VENDEDOR'] = endereco['bairro']
        df_cep_ven.at[idx, 'CIDADE_VENDEDOR'] = endereco['localidade']
        df_cep_ven.at[idx, 'ESTADO_VENDEDOR'] = endereco['uf']

    else:
        df_cep_ven.at[idx, 'LOGRADOURO_VENDEDOR'] = 'Nao Localizado'
        df_cep_ven.at[idx, 'BAIRRO_VENDEDOR'] = 'Nao Localizado'
        df_cep_ven.at[idx, 'CIDADE_VENDEDOR'] = 'Nao Localizado'
        df_cep_ven.at[idx, 'ESTADO_VENDEDOR'] = 'Nao Localizado'
```

```python
df_cep_ven.head()
```

```python
for idx, row in df_cep_cli.iterrows():
    cep = row.CEP_CLIENTE
    url = 'https://viacep.com.br/ws/%s/json/' % cep
    headers = {'User-Agent':'Teste'}
    requisicao = urllib.request.Request(url, headers=headers, method='GET')
    cliente = urllib.request.urlopen(requisicao)
    conteudo = cliente.read().decode('utf-8')
    endereco = json.loads(conteudo)
    cliente.close()
    print(cep)

    if 'cep' in endereco:
        df_cep_cli.at[idx, 'LOGRADOURO_CLIENTE'] = endereco['logradouro']
        df_cep_cli.at[idx, 'BAIRRO_CLIENTE'] = endereco['bairro']
        df_cep_cli.at[idx, 'CIDADE_CLIENTE'] = endereco['localidade']
        df_cep_cli.at[idx, 'ESTADO_CLIENTE'] = endereco['uf']

    else:
        df_cep_cli.at[idx, 'LOGRADOURO_CLIENTE'] = 'Nao Localizado'
        df_cep_cli.at[idx, 'BAIRRO_CLIENTE'] = 'Nao Localizado'
        df_cep_cli.at[idx, 'CIDADE_CLIENTE'] = 'Nao Localizado'
        df_cep_cli.at[idx, 'ESTADO_CLIENTE'] = 'Nao Localizado'
```

```python
df_cep_cli.head()
```

```python
df_cep_ven['BAIRRO_CIDADE'] = df_cep_ven['BAIRRO_VENDEDOR']+'-'+df_cep_ven['CIDADE_VENDEDOR']
df_cep_cli['BAIRRO_CIDADE'] = df_cep_cli['BAIRRO_CLIENTE']+'-'+df_cep_cli['CIDADE_CLIENTE']
```

```python
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
import geopy.distance

locator = Nominatim(user_agent='localizacao')
geocode = RateLimiter(locator.geocode, min_delay_seconds=1)
df_cep_ven['BAIRRO_CIDADE_GEOLOCALIZACAO'] = df_cep_ven['BAIRRO_CIDADE'].apply(geocode)
```

```python
df_cep_ven['BAIRRO_CIDADE_GEOLOCALIZACAO'].isna().sum()
```

```python
df_cep_ven.dropna(inplace=True)
```

```python
df_cep_ven['LATITUDE_VENDEDOR'] = df_cep_ven['BAIRRO_CIDADE_GEOLOCALIZACAO'].apply(lambda x: x.latitude)
df_cep_ven['LONGITUDE_VENDEDOR'] = df_cep_ven['BAIRRO_CIDADE_GEOLOCALIZACAO'].apply(lambda x: x.longitude)
df_cep_ven.drop('BAIRRO_CIDADE', axis=1, inplace=True)
df_cep_ven.drop('BAIRRO_CIDADE_GEOLOCALIZACAO', axis=1, inplace=True)
df_cep_ven
```

```python
locator = Nominatim(user_agent='localizacao')
geocode = RateLimiter(locator.geocode, min_delay_seconds=1)
df_cep_cli['BAIRRO_CIDADE_GEOLOCALIZACAO'] = df_cep_cli['BAIRRO_CIDADE'].apply(geocode)
df_cep_cli.head()
```

```python
df_cep_cli['BAIRRO_CIDADE_GEOLOCALIZACAO'].isna().sum()
```

```python
df_cep_cli.dropna(inplace=True)
```

```python
df_cep_cli['LATITUDE_CLIENTE'] = df_cep_cli['BAIRRO_CIDADE_GEOLOCALIZACAO'].apply(lambda x: x.latitude)
df_cep_cli['LONGITUDE_CLIENTE'] = df_cep_cli['BAIRRO_CIDADE_GEOLOCALIZACAO'].apply(lambda x: x.longitude)
df_cep_cli.drop('BAIRRO_CIDADE', axis=1, inplace=True)
df_cep_cli.drop('BAIRRO_CIDADE_GEOLOCALIZACAO', axis=1, inplace=True)
df_cep_cli.head()
```

```python
df_tratado = pd.merge(df, df_cep_ven, how='inner', on='CEP_VENDEDOR')
df_tratado = pd.merge(df_tratado, df_cep_cli, how='inner', on='CEP_CLIENTE')
df_tratado.head()
```

```python
for idx, row in df_tratado.iterrows():
    coord_vendedor = (row.LATITUDE_VENDEDOR, row.LONGITUDE_VENDEDOR)
    coord_cliente = (row.LATITUDE_CLIENTE, row.LONGITUDE_CLIENTE)

    df_tratado.at[idx, 'DISTANCIA_KM_VENDEDOR_CLIENTE'] = round(geopy.distance.distance(coord_vendedor, coord_cliente).km,2)
```
